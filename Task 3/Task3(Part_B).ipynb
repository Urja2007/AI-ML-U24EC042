```python
from sklearn.neighbors import NearestNeighbors

n_neighbors = 5  # Number of nearest neighbors for synthetic generation
num_augmented = 20  # Number of synthetic samples to create

nn = NearestNeighbors(n_neighbors=n_neighbors) #Initialize the KNN model
nn.fit(x)   #Fit on the original feature data (study hours)

augmented_x = [] #List to store synthetic inputs (study hours)
augmented_y = [] #List to store synthetic outputs (scores)

np.random.seed(42)
for i in range(num_augmented):
    idx = np.random.randint(0, len(x)) # Randomly pick one data point index
    x_i = x[idx] # Get the input (Hours studied)
    y_i = y[idx] #Get the corresponding output (Score))

    neighbors = nn.kneighbors([x_i], return_distance=False)#  Find indices of nearest neighbors of x_i
    neighbor_idx = np.random.choice(neighbors[0][1:])  # Randomly select **one** neighbor (excluding self)

    x_j = x[neighbor_idx] #  Get the neighbors input (study hours)
    y_j = y[neighbor_idx] # Get the neighbors output (score)

    # Linear interpolation
    lam = np.random.uniform(0, 1)
    synthetic_x = x_i + lam * (x_j - x_i) # Interpolated study hours
    synthetic_y = y_i + lam * (y_j - y_i) # Interpolated score

    augmented_x.append(synthetic_x[0]) # Store synthetic input (flattened)
    augmented_y.append(synthetic_y)#Store synthetic output
```

```python
# Combine original and synthetic input data
x_all = np.vstack([x, np.array(augmented_x).reshape(-1, 1)])
#  Combine original and synthetic output labels (scores)
y_all = np.concatenate([y, augmented_y])

#why both are not done by same method

# Re-train model
lr_aug = LinearRegression()
lr_aug.fit(x_all, y_all)
```
